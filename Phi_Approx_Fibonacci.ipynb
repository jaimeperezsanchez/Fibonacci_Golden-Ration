#### Import Decimal module (for extended precision decimals) and Matplotlib (for plotting)

```python
from decimal import *
import matplotlib.pyplot as plt 
```

### Golden Ration [Link](https://oeis.org/A001622)

```python
Phi = (1 + Decimal(5).sqrt())/2
print(Decimal(Phi))
```

### Fibonacci Sequence function [Link](https://www.geeksforgeeks.org/python-program-for-program-for-fibonacci-numbers-2/)

```python
def Fibonacci(n):
    a = 0
    b = 1
    if n < 0:
        return print("Incorrect number")
    elif n == 0:
        return a
    elif n == 1:
        return b
    else:
        for i in range(1,n):
            c = a + b
            a = b
            b = c    
        return b
```

#### Test Fibonacci Sequence function

```python
n_test = 10
print(Fibonacci(n_test))
```

### Approximation function of the Golden Ratio with the Fibonacci Sequence

```python
def Phi_approach(n):
    a = Decimal(Fibonacci(n+1)) / Decimal(Fibonacci(n))
    return a
```

#### Test Aproximation function

```python
n_test = 15
print(Phi_approach(n_test))
```

### Loop to test the approximations

```python
N_approx = 15
Phi_approx_sequence = []
Phi_errors = []
Phi_errors_abs = []

for i in range (1,N_approx):
    a = Phi_approach(i)
    Phi_approx_sequence.append(a)
    error = Decimal(Phi) - Decimal(a)
    Phi_errors.append(error)
    error_abs = abs(Decimal(Phi) - Decimal(a))
    Phi_errors_abs.append(error_abs)
```

#### Plot Phi Approximations

```python
Phi_approx_sequence
```

```python
plt.figure(figsize = (18, 12))
plt.plot(Phi_approx_sequence)
plt.title('Approximations to Phi')
plt.grid(True)
plt.show
```

#### Plot Phi Approximation Errors

```python
Phi_errors
```

```python
plt.figure(figsize = (18, 12))
plt.plot(Phi_errors)
plt.title('Approximation Errors to Phi')
plt.grid(True)
plt.show()
```
